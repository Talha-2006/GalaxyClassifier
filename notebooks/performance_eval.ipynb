```python
import torch
import matplotlib.pyplot as plt
```

```python
# Load the saved history
history = torch.load('../training_history.pt')
train_losses = history['train_losses']
val_losses = history['val_losses']
train_accs = history['train_accs']
val_accs = history['val_accs']

epochs = range(1, len(train_losses) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(epochs, train_losses, label='Train Loss', marker='o')
ax1.plot(epochs, val_losses, label='Val Loss', marker='o')
ax1.axvline(x=20, color='red', linestyle='--', alpha=0.5, label='LR reduced')
ax1.axvline(x=26, color='orange', linestyle='--', alpha=0.5, label='LR reduced again')
ax1.set_title('Loss over Epochs')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

ax2.plot(epochs, train_accs, label='Train Acc', marker='o')
ax2.plot(epochs, val_accs, label='Val Acc', marker='o')
ax2.axvline(x=20, color='red', linestyle='--', alpha=0.5, label='LR reduced')
ax2.axvline(x=26, color='orange', linestyle='--', alpha=0.5, label='LR reduced again')
ax2.set_title('Accuracy over Epochs')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig('assets/learning_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved.")
```
